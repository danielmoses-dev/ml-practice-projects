```python
import pandas as pd
import numpy as np
```

```python
#reading and storing datasets
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
#merging the two datasets
movies = movies.merge(credits, on='title')
```

```python
movies.head()
```

```python
#removing unwanted columns
#what we keep: movie_id, title, overview, genre, cast, crew

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
#check for nulls
movies.isna().sum()
```

```python
#remove nulls
movies.dropna(inplace = True)
```

```python
#check for duplicates
movies.duplicated().sum()
```

```python
#clean up genres
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):        #convert string to list
        L.append(i['name'])
    return L
```

```python
#use apply() to call the function we created and apply it on the dataset
movies['genres'] = movies.genres.apply(convert)
```

```python
#use same function on keywords column
movies['keywords'] = movies.keywords.apply(convert)
```

```python
#modify our function for cast to get first three 
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
#get director only from cast
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
#apply to the crew
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#convert overview from string to list using split
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#remove spaces from names to make it a single tag
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
#check dataset
movies.head()
```

```python
#combine and create tags column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#create new dataset to include only necessary stuff
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
#convert the tags list into a string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#convert all tags to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
#VECTORIZATION
#take 5000 most frequent words 
#and then calculate the frequency of these for each movie
#however we will not consider stop words i.e words like and, it, because, etc

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#display all 5000 top words

cv.get_feature_names()
```

```python
#library

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#stemming i.e getting root of each word eg: lov for love, loves, loving, loved

def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#now go up and again run the nltk cells

#now we will find the cosine distance (similarity) of all 5000 movies for each of the 5000 movies

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
#create the main function which takes the movie
#however we need to sort the similarity in descending to get the closest vectors
#but we will lose the order of vectors and we wont know which movie the vector belongs to

#we are not enumerating everything. 
#these things are done for the given movie only

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
